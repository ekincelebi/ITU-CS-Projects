```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 5GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session

test_t0 = pd.read_csv("test_t0.csv")
train_t0 = pd.read_csv("train_t0.csv")
train_t1 = pd.read_csv("train_t1.csv")
#ece is now currently testing her git skills

import statsmodels.api as sm
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import math
def labelspre(y):
    y = np.array(y).copy()
    epsilon = 1e-100
    for i in range(y.shape[0]):
        if(y[i] > 1):
            y[i] = 1
        if(y[i] < 0):
            y[i] = 0
        if((1-y[i]) < epsilon):
            y[i] = 1e+200
        elif(y[i] < epsilon):
            y[i] = -1e+200
        else:
            temp = y[i]
            y[i] = np.log(temp/(1-temp))
    return y

def labelspost(y):
    y = np.array(y)
    for i in range(y.shape[0]):
        for j in range(y.shape[1]):
            y[i][j] = 1/(1 + np.exp(-y[i][j]))
    return y 
    
class MultioutputLearner:
    def __init__(self,X,Y):
        epsilon = 1e-100
        X = np.array(X)
        self.X = X.copy()
        self.Y = np.delete(np.array(Y).copy(), [0], axis=1)
        self.listofLR = []
        
    def train(self):
        pca = PCA(n_components=80)
        temp_X = pca.fit_transform(self.X)
        for i in range(self.Y.shape[1]):
            temp_y = labelspre(self.Y[:,i]).copy()
            temp_model = LinearRegression().fit(temp_X,temp_y)
            self.listofLR.append(temp_model)

    def test(self,testset):
        testoutput = []
        testset = np.array(testset)
        pca = PCA(n_components=80)
        temp_test = pca.fit_transform(testset)
        for LR in self.listofLR:
            prediction = LR.predict(temp_test).copy()
            testoutput.append(prediction)
        return np.transpose(labelspost(np.array(testoutput)))
                
model = MultioutputLearner(train_t0,train_t1)
model.train()

testresult = model.test(test_t0).flatten()
print(testresult)
samplesubmission = pd.read_csv('sampleSubmission.csv')

output = pd.DataFrame({'ID': samplesubmission.index, 'Predicted': testresult})
output.to_csv('submissiondeneme2.csv', index=False)
```
